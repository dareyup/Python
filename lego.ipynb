- Create a function called `set_parts` that takes as input the set number (as a string) and version number (as an integer) and outputs a DataFrame with each part in the set.  
- The DataFrame should have the following columns/column names: `set_num`, `version`, `name_sets`, `year`, `name_themes`, `part_num`, `name_part`, `name_color`,`quantity`,`is_spare`.  
- The Data should be sorted by `part_num` and then `name_color`.  
- The index should go from 0 to n-1.

```python
# join iv parts on parts then join on colors -- two left join 
pawts = pd.merge(inventory_parts, parts)

pawts = pawts.rename(columns = {'name':'name_part'})

p2 = pd.merge(pawts, colors, how = 'outer', left_on = 'color_id', right_on = 'id')

# merge sets with inventories to get version
d3 = d2.merge(inventories, on = 'set_num')
d3 = d3[['set_num', 'version', 'name_sets', 'year', 'name_themes', 'id']]

# merge d3 and p2 somehow * insert smiley face *
full_merge = pd.merge(d3, p2, how = 'outer', left_on = 'id', right_on = 'inventory_id')

full_merge = full_merge.rename(columns = {'name':'name_color'})

# subset columsn 
full_merge = full_merge[['set_num', 'version', 'name_sets', 'year', 'name_themes', 'part_num', 
           'name_part', 'name_color', 'quantity', 'is_spare']]

full_merge['year'] = full_merge['year'].fillna(0).astype(int)
full_merge['version'] = full_merge['version'].fillna(0).astype(int)
full_merge['quantity'] = full_merge['quantity'].fillna(0).astype(int)

def set_parts(set_num, vers):
    return full_merge[(full_merge.version == vers) & (full_merge.set_num == set_num)].sort_values(by = ['part_num', 'name_color']).reset_index(drop = True)
```

**Q25:**  
- Create a function called `get_part_colors` that takes as input a part number (as a string) and returns a DataFrame of all colors a part has appeared in.  
- The DataFrame should be sorted by color id  
- The index should go from 0 to n-1.

```python
# merge colors and elemets tables 
p3 = pd.merge(colors, elements, how = 'left', left_on = 'id', right_on = 'color_id')

# create function 
def get_part_colors(part_nums):
    return p3[p3.part_num == part_nums].sort_values('color_id').drop_duplicates('name').reset_index(drop = True)[['id', 'name', 'rgb', 'is_trans']]
```
